```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
en_dir="energy_new"

energies = np.load(f"{en_dir}/en_expl.npy")
times = np.load(f"{en_dir}/times_expl.npy")
```

```python
plt.clf()

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(times, energies[:, 0], label="expl")
ax2.plot(times, energies[:, 1], label="expl", marker="o")

for i in [1]:#[1, 2, 4, 8, 10, 15, 20, 30, 50]:
    energies = np.load(f"{en_dir}/en_impl_{i}.npy")
    times = np.load(f"{en_dir}/times_impl_{i}.npy")

    print(energies)

    ax1.plot(times, energies[:, 0], label=f"impl {i}")
    ax2.plot(times, energies[:, 1], label=f"impl {i}")

ax1.legend()
ax2.legend()
```
